```python
#importing the relevant packages
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline
from Bio import AlignIO
import warnings
warnings.filterwarnings("ignore")

#function to calculate the Shannon Entropy of a MSA
# H = -sum[p(x).log2(px)]

def shannon_entropy(list_input):
    unique_aa = set(list_input)   
    M = len(list_input)
    entropy_list = []
    # Number of residues in column
    for aa in unique_aa:
        n_i = list_input.count(aa)                                           
        P_i = n_i/float(M)                                
        entropy_i = P_i*(math.log(P_i,10))
        entropy_list.append(entropy_i)
    sh_entropy = -(sum(entropy_list))
    #print(sh_entropy)
    return sh_entropy

#importing the MSA file
#importing the clustal file
align_clustal1 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_nov2020_clustal.aln", "clustal")
align_clustal2 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_dec2020_clustal.aln", "clustal")
align_clustal3 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_jan2021_clustal.aln", "clustal")
align_clustal4 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_feb2021_clustal.aln", "clustal")
align_clustal5 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_mar2021_clustal.aln", "clustal")
align_clustal6 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_apr2021_clustal.aln","clustal")
align_clustal7 = AlignIO.read("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/Brazil/Gamma/gamma_may2021_clustal.aln","clustal")

def shannon_entropy_list_msa(alignment_file):
    shannon_entropy_list = []
    for col_no in range(len(list(alignment_file[0]))):
        list_input = list(alignment_file[:, col_no])
        shannon_entropy_list.append(shannon_entropy(list_input))
    return shannon_entropy_list

clustal_omega1 = shannon_entropy_list_msa(align_clustal1)
clustal_omega2 = shannon_entropy_list_msa(align_clustal2)
clustal_omega3 = shannon_entropy_list_msa(align_clustal3)
clustal_omega4 = shannon_entropy_list_msa(align_clustal4)
clustal_omega5 = shannon_entropy_list_msa(align_clustal5)
clustal_omega6 = shannon_entropy_list_msa(align_clustal6)
clustal_omega7 = shannon_entropy_list_msa(align_clustal7)

# Plotting the data
fig, axs = plt.subplots(7, sharex=True, sharey=True, figsize=(18,20))
axs[0].spines['right'].set_color(None)
axs[0].spines['top'].set_color(None)
axs[0].spines['left'].set_linewidth(2)
axs[0].spines['bottom'].set_linewidth(2)
axs[1].spines['right'].set_color(None)
axs[1].spines['top'].set_color(None)
axs[1].spines['left'].set_linewidth(2)
axs[1].spines['bottom'].set_linewidth(2)
axs[2].spines['right'].set_color(None)
axs[2].spines['top'].set_color(None)
axs[2].spines['left'].set_linewidth(2)
axs[2].spines['bottom'].set_linewidth(2)
axs[3].spines['right'].set_color(None)
axs[3].spines['top'].set_color(None)
axs[3].spines['left'].set_linewidth(2)
axs[3].spines['bottom'].set_linewidth(2)
axs[4].spines['right'].set_color(None)
axs[4].spines['top'].set_color(None)
axs[4].spines['left'].set_linewidth(2)
axs[4].spines['bottom'].set_linewidth(2)
axs[5].spines['right'].set_color(None)
axs[5].spines['top'].set_color(None)
axs[5].spines['left'].set_linewidth(2)
axs[5].spines['bottom'].set_linewidth(2)
axs[6].spines['right'].set_color(None)
axs[6].spines['top'].set_color(None)
axs[6].spines['left'].set_linewidth(2)
axs[6].spines['bottom'].set_linewidth(2)
axs[0].plot(clustal_omega1, 'black', linewidth=2.0)
axs[1].plot(clustal_omega2, 'red', linewidth=2.0)
axs[2].plot(clustal_omega3, 'blue', linewidth=2.0)
axs[3].plot(clustal_omega4, 'magenta', linewidth=2.0)
axs[4].plot(clustal_omega5, 'darkgreen', linewidth=2.0)
axs[5].plot(clustal_omega6, 'pink', linewidth=2.0)
axs[6].plot(clustal_omega7, 'cyan', linewidth=2.0)
axs[0].tick_params(axis='y', labelsize=20)
axs[1].tick_params(axis='y', labelsize=20)
axs[2].tick_params(axis='y', labelsize=20)
axs[3].tick_params(axis='y', labelsize=20)
axs[4].tick_params(axis='y', labelsize=20)
axs[5].tick_params(axis='y', labelsize=20)
axs[6].tick_params(axis='y', labelsize=20)
axs[6].tick_params(axis='x', labelsize=20)
plt.ylim(0,0.4)
plt.xlim(21563, 25385)
#plt.savefig("/home/satyam/Documents/Projects/Entropy_Comparison/GISAID/India/Updated_result_27dec/delta_combine_fluctuation.png", dpi=300)
plt.show()
```
