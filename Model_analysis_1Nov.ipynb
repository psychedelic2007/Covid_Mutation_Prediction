```python
import numpy as np
import matplotlib.pyplot as plt

accuracy_mean = 98.59954079
precision_mean = 99.72158393
recall_mean = 98.87171736
f1_score_mean = 99.29481327

accuracy_std = 0.087392491
precision_std = 0.088387
recall_std = 0.000988781
f1_score_std = 0.044315237

# Create lists for the plot
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
x_pos = np.arange(len(metrics))
CTEs = [accuracy_mean,precision_mean,recall_mean,f1_score_mean]
error = [accuracy_std,precision_std,recall_std,f1_score_std]

# Build the plot
fig, ax = plt.subplots(figsize=(18,10))
ax.bar(x_pos, CTEs, color='grey',yerr=error, align='center', alpha=1, edgecolor='black',capsize=20, 
       width=0.3, linewidth=3.0, error_kw=dict(ecolor='green',elinewidth=0.5, capthick=3))
ax.set_ylabel('Percentage', fontsize=30)
ax.set_xticks(x_pos)
ax.set_xticklabels(metrics, fontsize=30)
plt.yticks(fontsize=25)
ax.set_title('Model Metrics', fontsize=30)
plt.ylim(98,100)
# Save the figure and show
plt.tight_layout()
plt.savefig('/home/satyam/Documents/Projects/Final_manuscript/Figures/Supplementary/model_metric2.png', dpi=300)
plt.show()
```
